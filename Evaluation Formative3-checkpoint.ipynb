```python
class CompteBancaire:
    def __init__(self,numeroCompte,nomProprietaire,solde,dateOuvertureCompte,nomAgenceOuvertureCompte,typeDeCompte,depotInitial):
        self.numeroCompte = numeroCompte
        self.nomProprietaire = nomProprietaire
        self.solde = solde
        self.dateOuvertureCompte = dateOuvertureCompte
        self.nomAgenceOuvertureCompte = nomAgenceOuvertureCompte
        self.typeDeCompte = typeDeCompte
        self.depotInitial = depotInitial
    def versement(self, montant):
        if self.solde == depotInitial:
            self.solde += depotInitial
        else:
            self.solde += montant
            
    def retrait(self,montant):
        if self.solde <= self.montant:
            print("Solde insuffisant\n Operation non autorisée")
        else :
            self.solde -= montant
    def agios(self):
        self.solde *= 0.95
    def afficher(self):
        print("Affichage des informations du compte")
        print(self.numeroCompte, " est le numéro de votre compte")
        print("Ce compte est à la propriété de ",self.nomProprietaire)
        print("Le solde de votre compte est de ",self.solde,"€")
        print("Compte ouvert le ",self.dateOuvertureCompte)
        print("A l'agence ",self.nomAgenceOuvertureCompte)
        print("Compte de type ",self.typeDeCompte)
        print("Votre depôt initial est de ",self.depotInitial,"€")
        print("="*100)
    def saisie(self):
        print("Veuillez entrer les informations qui vous seront démandées!")
        self.numeroCompte = input("Entrez le numéro de votre compte :")
        self.nomProprietaire = input("Entrer votre nom complet :")
        self.solde = int(input("Entrer le montant de votre solde actuel :"))
        self.dateOuvertureCompte = input("Entrer la date -jj-mm-aa- d'ouverture de votre compte :")
        self.nomAgenceOuvertureCompte = input("nom de l'agence d'ouverture de votre compte :")
        self.typeDeCompte = input("Compte de type? :")
        self.depotInitial = int(input("Veuillez entrer le montant de votre depôt initial :"))
        print("="*100)
        
    # Créeons pour chaque propriété son getter et son setter
    # Création de getters
    def getNumeroCompte(self,numeroCompte):
        return self.numeroCompte
    def getNomProprietaire(self,nomProprietaire):
        return self.nomProprietaire
    def getSolde(self,solde):
        return self.solde
    def getDateOuvertureCompte(self,dateOuvertureCompte):
        return self.dateOuvertureCompte
    def getNomAgenceOuvertureCompte(self,nomAgenceOuvertureCompte):
        return self.nomAgenceOuvertureCompte
    def getTypeDecompte(self,typeDeCompte):
        return self.typeDeCompte
    def getDepotInitial(self,depotInitial):
        return self.depotInitial
    
    # Création des setters
    def setNumeroCompte(self, numeroCompte):
        self.numeroCompte = numeroCompte
    def setNomProprietaire(self, nomProprietaire):
        self.nomProprietaire = nomProprietaire
    def setSolde(self,solde):
        self.solde = solde
    def setDateOuvertureCompte(self, dateOuvertureCompte):
        self.dateOuvertureCompte = dateOuvertureCompte
    def setNomAgenceOuvertureCompte(self, nomAgenceOuvertureCompte):
        self.nomAgenceOuvertureCompte = nomAgenceOuvertureCompte
    def setTypeDeCompte(self,typeDeCompte):
        self.typeDeCompte = typeDeCompte
    def setDepotInitial(self,depotInitial):
        self.depotInitial = depotInitial
    
c1 = CompteBancaire('','',0,'','','',0)
c2 = CompteBancaire('','',0,'','','',0)
c3 = CompteBancaire('','',0,'','','',0)
c4 = CompteBancaire('','',0,'','','',0)
# Ajoutons les objets dans une liste   
liste = [c1,c2,c3,c4]
def afficherListe():
    for i in liste:
        i.saisie()
    for i in liste:
        i.afficher()
afficherListe()
            
        
            
       
```
